```python
# pip install folium
import folium
from folium.plugins import PolyLineTextPath
```

```python
# Beispiel-Daten
data = [
    {"lon1": 10, "lat1": 51, "lon2": 11, "lat2": 52},
    {"lon1": 11, "lat1": 52, "lon2": 12, "lat2": 53}
]
```

```python
m = folium.Map(location=[51.5, 11], zoom_start=6)

for row in data:
    points = [(row["lat1"], row["lon1"]), (row["lat2"], row["lon2"])]
    line = folium.PolyLine(points, color="blue", weight=1)
    m.add_child(line)
    # Pfeilspitze nur am Ende
    PolyLineTextPath(
        line,
        '➤',
        repeat=False,
        offset=5,
        attributes={'font-size': '15', 'fill': 'blue'}
    ).add_to(m)

m.save("karte_mit_pfeilen.html")
```

```python
import folium
import math

def bearing(p1, p2):
    lat1, lon1 = math.radians(p1[0]), math.radians(p1[1])
    lat2, lon2 = math.radians(p2[0]), math.radians(p2[1])
    dlon = lon2 - lon1
    x = math.sin(dlon) * math.cos(lat2)
    y = math.cos(lat1) * math.sin(lat2) - math.sin(lat1) * math.cos(lat2) * math.cos(dlon)
    return (math.degrees(math.atan2(x, y)) + 360) % 360

m = folium.Map(location=[51, 10], zoom_start=6)

lines = [
    [(52.5200, 13.4050), (53.5511, 9.9937)],
    [(48.1351, 11.5820), (50.1109, 8.6821)]
]

for start, end in lines:
    folium.PolyLine([start, end], color="blue", weight=3).add_to(m)

    angle = bearing(start, end)

    folium.Marker(
        location=end,
        icon=folium.DivIcon(
            html=f"""<div style="
                transform: rotate({angle}deg);
                font-size: 20px;
                color: blue;
            ">➤</div>"""
        )
    ).add_to(m)

m.save("karte_mit_pfeilen.html")
```

```python
import folium
import math

def bearing(p1, p2):
    lat1, lon1 = math.radians(p1[0]), math.radians(p1[1])
    lat2, lon2 = math.radians(p2[0]), math.radians(p2[1])
    dlon = lon2 - lon1
    x = math.sin(dlon) * math.cos(lat2)
    y = math.cos(lat1) * math.sin(lat2) - math.sin(lat1) * math.cos(lat2) * math.cos(dlon)
    return (math.degrees(math.atan2(x, y)) + 360) % 360

m = folium.Map(location=[51, 10], zoom_start=6)

lines = [
    [(52.5200, 13.4050), (53.5511, 9.9937)],
    [(48.1351, 11.5820), (50.1109, 8.6821)]
]

for start, end in lines:
    folium.PolyLine([start, end], color="blue", weight=3).add_to(m)
    angle = bearing(start, end)
    # SVG-Pfeil, Spitze exakt auf Endpunkt
    svg_arrow = f"""
    <svg width="24" height="24" style="transform: rotate({angle + 90}deg);">
      <polygon points="0,12 18,6 18,18" fill="blue"/>
    </svg>
    """
    folium.Marker(
        location=end,
        icon=folium.DivIcon(html=f"""<div style="transform: translate(0px, -0px);">{svg_arrow}</div>""")
    ).add_to(m)

m.save("karte_mit_pfeilen.html")
```
